```python
#Load data from per-capita-energy-use.csv

data = pd.read_csv('..//data//per-capita-energy-use.csv')
df = data.copy()
```

CONSUMO PER CAPITA: Como consumo per cápita se conoce el consumo total (de productos, de alimentos, de agua, de energía, etc.) de un país o región dividido por el número de sus habitantes en determinado periodo de tiempo.

```python
# Lista de países a mantener
paises_a_mantener = ['Austria', 'Belgium', 'Bulgaria', 'Croatia', 'Cyprus', 'Czech Republic', 'Denmark', 'Estonia', 'Europe', 'Finland',
   'France', 'Germany', 'Greece', 'Greenland',
   'Hungary', 'Iceland', 'Ireland', 'Italy', 'Latvia', 'Lithuania',
   'Luxembourg', 'Malta', 'Netherlands', 'Norway', 'Poland', 'Portugal', 'Romania',
   'Slovakia', 'Slovenia', 'Spain', 'Sweden', 'Switzerland','United Kingdom', 'World']

# Eliminar filas que no están en la lista de paises_a_eliminar
df_filtrado = df[df['Entity'].isin(paises_a_mantener)]

# El DataFrame df_filtrado ahora contiene solo las filas correspondientes a los países en paises_a_eliminar
print(df_filtrado)
```

```python
df.shape
```

```python
df = df_filtrado
df.reset_index(inplace=True, drop=True) #RESETEAR INDICE 
```

```python
df = df.rename(columns={'Entity': 'Country'}) #RENOMBRAR COLUMNA 
```
